```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
meu = np.array([3,2])
sigma = np.array([
    [10,0],
    [0,1]
])
```

```python
multi_norm = np.random.multivariate_normal(meu,sigma,1000)
```

```python
multi_norm.shape
```

```python
#multi_norm[row_slicing,column_slicing]
```

```python
x=multi_norm[:,0]
y=multi_norm[:,1]
```

```python
plt.scatter(x,y)
plt.title('title')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
plt.hist2d(x,y)
plt.show()
```

```python
np.corrcoef(x,y)
```
