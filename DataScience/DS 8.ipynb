```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
y1= np.linspace(10,100,15)
y2= y1**2
y3= y1**3
```

```python
x=np.linspace(1,15,15)
```

```python
plt.figure(figsize=(20,10))

plt.title('My plot')
plt.xlabel('Speed')
plt.ylabel('Distance')
plt.plot(x,y1,color='r',label='y1(red)',marker='x')
plt.plot(x,y2,color='b',label='y2(blue)',marker='*')
plt.plot(x,y3,color='g',label='y3(green)',marker='+' )
plt.legend()
plt.show()
```

```python
labels =['A','B','C']
data =(80,40,60)
explosion=(0,0.1,0)
custom_colours=('red','hotpink','#4DAF42')
plt.pie(x=data, labels=labels ,explode=explosion)
plt.show()
```

```python
plt.barh(labels,data,color=custom_colours)
plt.savefig('bar plt.png')
```
