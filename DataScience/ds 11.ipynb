```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering
```

```python
data = pd.read_csv('../dataset/Mall_Customers.csv')
```

```python
data.head()
```

```python
x=data[['Annual Income (k)','Spending Score (1-100)']]
```

```python
x.sample(5)
```

```python
x=data[['Annual Income (k)','Spending Score (1-100)']].copy()
```

```python
x.head()
```

```python
plt.scatter(x['Annual Income (k)'],x['Spending Score (1-100)'])
plt.xlabel('Annual Income(k)')
plt.ylabel('Spending Score(1-100)')
plt.show()
```

```python
sse = []#sum of squared error
for clusters in range(1,10):
    K = KMeans(n_clusters=clusters)
    K.fit(x)
    sse.append(K.inertia_)
```

```python
sse
```

```python
plt.plot(range(1,10),sse)
plt.xlabel('clusters(K)')
plt.ylabel('sum of Squared Error (sse)')
plt.title('Elbow plot')
plt.show()
```

```python
Kmeans = KMeans(n_clusters=5,random_state=666)
Kmeans.fit(x)
```

```python
Kmeans.labels_
```

```python
x
```

```python
x = x.values
```

```python
cluster_labels = Kmeans.labels_
```

```python
plt.scatter(x[cluster_labels==0,0],x[cluster_labels==0,1],c='r',label='cluster 0')
plt.scatter(x[cluster_labels==1,0],x[cluster_labels==1,1],c='g',label='cluster 1')
plt.scatter(x[cluster_labels==2,0],x[cluster_labels==2,1],c='b',label='cluster 2')
plt.scatter(x[cluster_labels==3,0],x[cluster_labels==3,1],c='y',label='cluster 3')      
plt.scatter(x[cluster_labels==4,0],x[cluster_labels==4,1],c='r',label='cluster 4')      
plt.xlabel('annual income')
plt.ylabel('spending score')
plt.legend()
plt.show()
```

```python
pred_data=[[90,4],[120,18],[20,90]]
```

```python
Kmeans.predict(pred_data)
```

```python
agg=AgglomerativeClustering(n_clusters=5)
agg.fit(x)
cluster_labels=agg.labels_
```

```python
plt.scatter(x[cluster_labels==0,0],x[cluster_labels==0,1],c='r',label='cluster 0')
plt.scatter(x[cluster_labels==1,0],x[cluster_labels==1,1],c='g',label='cluster 1')
plt.scatter(x[cluster_labels==2,0],x[cluster_labels==2,1],c='b',label='cluster 2')
plt.scatter(x[cluster_labels==3,0],x[cluster_labels==3,1],c='y',label='cluster 3')      
plt.scatter(x[cluster_labels==4,0],x[cluster_labels==4,1],c='k',label='cluster 4')      
plt.xlabel('annual income')
plt.ylabel('spending score')
plt.legend()
plt.show()
```
