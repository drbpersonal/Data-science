```python
import torch
import cv2
```

```python
from ultralytics import YOLO
```

```python
model = YOLO('../yolov8n.pt')
```

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
device
```

```python
model.to(device)
```

```python
classes = model.names
classes
```

```python
# cap = cv2.VideoCapture(0)
cap = cv2.VideoCapture('../Road traffic video for object recognition.mp4')
fourcc = cv2.VideoWriter_fourcc(*'MJPG')
out = cv2.VideoWriter('video.mp4', fourcc, 24, (1280, 720))

while cap.isOpened():
    success, img = cap.read()
    if not success:
        break
    width = img.shape[1]
    height = img.shape[0]

    results = model(img)
    for result in results:
        objects = result.boxes.cls
        bboxes = result.boxes.xyxyn
        confidences = result.boxes.conf
        for obj, conf, box in zip(objects, confidences, bboxes):
            conf = conf.tolist()
            if conf < 0.5:
                continue
            conf = round(conf, 1)
            x1, y1 = int(box[0] * width), int(box[1] * height)
            x2, y2 = int(box[2] * width), int(box[3] * height)
            label = classes[obj.tolist()]
            cv2.rectangle(img, (x1, y1), (x2, y2), (0, 0, 255), 2)
            cv2.putText(img, f'{label}: ({conf})', (x1, y1+30),  cv2.FONT_HERSHEY_COMPLEX, 1, (125, 125, 125), 2)
    out.write(img)
    cv2.imshow('image', img)
    if cv2.waitKey(1) & 0xff == 27:
        cap.release()
        cap.destroyAllWindows()
        break
        
    
cap.release()
cv2.destroyAllWindows()
```
