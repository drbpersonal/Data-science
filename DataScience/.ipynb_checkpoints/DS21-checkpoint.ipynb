```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout
```

```python
from keras.optimizers import SGD
```

```python
from keras.datasets import mnist
```

```python
(X_train,y_train),(X_test,y_test)=mnist.load_data()
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
image_index =59999
plt.imshow(X_train[image_index],plt.cm.binary)
plt.title(y_train[image_index])
plt.show
```

```python
[
[1,5,8],
[8,7,6],
[9,3,5]
]
[1,5,8,8,7,6,9,3,5]
X_train = X_train.reshape(60_000,784)
X_test =X_test.reshape(10_000,784)
```

```python
np.unique(y_train)
```

```python
from collections import Counter
Counter(y_train)
```

```python
y_train [:5]
```

```python
from keras.utils import to_categorical
```

```python
y_train =to_categorical(y_train)
y_test = to_categorical(y_test)
y_train[:5]
```

```python
model=Sequential()

model.add(Dense(input_dim=784,units=64, activation='relu'))
model.add(Dropout(0.2))

#HIdden layer
model.add(Dense(units = 32,activation='relu'))
model.add(Dropout(0.2))

model.add(Dense(units = 16,activation='relu'))
model.add(Dropout(0.2))
#Output layer
model.add(Dense(units = 10,activation='softmax'))

                
                
model.compile(loss='categorical_crossentropy', optimizer=SGD(learning_rate=1e-4),metrics=['accuracy'])                
```

```python
1e-4
```

```python
model.summary()
```

```python
from keras.callbacks import EarlyStopping,ModelCheckpoint
```

```python
es=EarlyStopping(monitor='val_loss',patience=5)
checkpoint = ModelCheckpoint(filepath='./checkpoint.h5',
                            save_best_only=True,
                            save_weights_only=False)
```

```python
history = model.fit(X_train,y_train,
                   validation_data=(X_test,y_test),
                   epochs=5000,
                    callbacks=[es,checkpoint]
                   )
```
