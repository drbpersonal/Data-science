```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
```

```python
data = pd.read_csv('../dataset/heart.csv')
```

```python
data.sample(5)
```

```python
to_drop = ['fbs', 'restecg', 'exang']
data.drop(to_drop, axis=1, inplace=True)
```

```python
data.describe()
```

```python
data['ca'].unique()
```

```python
data['oldpeak'].unique()
```

```python
data.isna().sum()
```

```python
data.dropna(inplace=True)
```

```python
data.shape
```

```python
data.drop_duplicates(inplace=True)
```

```python
data.shape
```

```python
y = data['target']
X = data.drop(['target'], axis=1)
```

```python
from collections import Counter
Counter(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.15,
                                                    random_state=666,
                                                    stratify=y)
```

```python
dtree = DecisionTreeClassifier()
rand_forest = RandomForestClassifier(n_jobs=-1)
ada_boost = AdaBoostClassifier(learning_rate=0.1)
```

```python
%%time
dtree.fit(X_train, y_train)
```

```python
features = list(X_train.columns)
features
```

```python
plt.figure(figsize=(50, 50))
plot_tree(dtree, feature_names=features, filled=True)
plt.show()
```

```python
rand_forest.fit(X_train, y_train)
ada_boost.fit(X_train, y_train)
```

```python
features
```

```python
feature_imp = dtree.feature_importances_
feature_imp
```

```python
for imp, feat in zip(feature_imp, features):
    print(f'{feat}: {imp * 100}')
```

```python
y_pred = dtree.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
feature_imp = rand_forest.feature_importances_
for imp, feat in zip(feature_imp, features):
    print(f'{feat}: {imp * 100}')
```

```python
y_pred = rand_forest.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
feature_imp = ada_boost.feature_importances_
for imp, feat in zip(feature_imp, features):
    print(f'{feat}: {imp * 100}')
```

```python
y_pred = ada_boost.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic = LogisticRegression(n_jobs=-1,)
```

```python
logistic.fit(X_train, y_train)
```

```python
y_pred = logistic.predict(X_test)
```

```python
print(classification_report(y_test, y_pred))
```
