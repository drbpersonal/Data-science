```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from sklearn.linear_model import LinearRegression
from scipy.stats import linregress
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

* Mean Absolute Error(MAE) = $\frac{\sum|y_{actual} - y_{predicted}|}{N}$
* Mean Squared Error(MSE) = $\frac{\sum(y_{actual} - y_{predicted})^2}{N}$
* R2 = $1 - \frac{\sum(y_{actual} - y_{predicted})^2}{\sum(y_{actual} - y_{mean})^2}$

```python
X = np.random.normal(10, 2, 120)
y = X + np.random.normal(3, 0.5, 120)
```

```python
X[:5]
```

```python
y[:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
 X_train, X_test, y_train, y_test  = train_test_split(X, y,
                                                      test_size=0.2,
                                                      random_state=666,
                                                      shuffle=True,)
```

```python
plt.scatter(X_train, y_train, c='b', label='Training Data')
plt.scatter(X_test, y_test, c='r', label='Testing Data')
plt.legend()
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
m, c, r2, _, err = linregress(X_train, y_train)
```

```python
reg_line_train = m * X_train + c
```

```python
plt.scatter(X_train, y_train)
plt.plot(X_train, reg_line_train, c='r')
plt.show()
```

```python
reg_line_test =m*X_test+c
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, reg_line_test, c='r')
plt.show()
```

```python
def predict(x_new):
    return m * x_new + c
```
